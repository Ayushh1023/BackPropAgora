```python
import math
import numpy as np
import matplotlib.pyplot as plt 
```

```python
def f(x):
    return 4*x**2 - 2*x + 5
```

```python
f(3.0), f(-4.0)
```

```python
x = np.arange(-10, 10, 0.25)
y = f(x)
plt.plot(x,y)
plt.grid()
```

```python
h = 0.001
x = 3.0

(f(x+h)-f(x))/h
```

```python
x = -3.0

(f(x+h)-f(x))/h
```

```python
h = 0.0001

a = 2.0
b = -3.0
c = 10.0
d1 = a*b + c
a+=h
d2 = a*b + c

print('d1', d1)
print('d2', d2)
print('slope', (d2-d1)/h)
```
